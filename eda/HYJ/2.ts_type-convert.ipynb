```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
test = pd.read_parquet("/kaggle/input/otto-full-optimized-memory-footprint/test.parquet")
train = pd.read_parquet("/kaggle/input/otto-full-optimized-memory-footprint/train.parquet")
```

```python
!pip install pickle5

import pickle5 as pickle

with open('../input/otto-full-optimized-memory-footprint/id2type.pkl', "rb") as fh:
    id2type = pickle.load(fh)
with open('../input/otto-full-optimized-memory-footprint/type2id.pkl', "rb") as fh:
    type2id = pickle.load(fh)
```

```python
# type 돌려놓기

train["type"] = train["type"].map(lambda i: id2type[i])
test["type"] = test["type"].map(lambda i: id2type[i])
```

```python
# ts 전처리

### train
train.ts = pd.to_datetime(train.ts * 1e9)
train['day'] = train.ts.dt.day
train['hour'] = train.ts.dt.hour
print('Train min date and max date are:', train.ts.min(),'and', train.ts.max() )

### test
test.ts = pd.to_datetime(test.ts * 1e9)
test['day'] = test.ts.dt.day
test['hour'] = test.ts.dt.hour
print('Test min date and max date are:', test.ts.min(),'and', test.ts.max() )


# print('We will truncate train data to begin Aug 1st, 2022')
# train = train.loc[train.ts >= cudf.to_datetime('2022-08-01')]
```

```python
train.day = train.day.astype("int16")
test.day = test.day.astype("int16")
train.hour = train.hour.astype("int16")
test.hour = test.hour.astype("int16")
```

```python
train.to_parquet("train.parquet")
test.to_parquet("test.parquet")
```
